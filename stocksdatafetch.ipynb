```python
import yfinance as yf
import pandas as pd

# Define the stock ticker symbol for Tata Motors
ticker_symbol = 'TATAMOTORS.NS'

# Download historical data
# period1 = start date (in seconds since epoch)
# period2 = end date (in seconds since epoch)
# In this case, period1 = January 1, 1991 and period2 = October 12, 2024
tatamotors_data = yf.download(ticker_symbol, start='1991-01-01', end='2024-10-12', interval='1mo')

# Save the data to a CSV file
csv_filename = 'tatamotors_historical_data.csv'
tatamotors_data.to_csv(csv_filename)

print(f"Data downloaded and saved to {csv_filename}")


print("safchkbshd")
```
